```python
#Recode wave into months based on "Speek Dating Data Key" word doc. 
def score_to_numeric(x):
    if x<=2:
        return 10
    if x==3:
        return 11
    if x==4:
        return 11
    if x==5:
        return 11
    if x==6:
        return 3
    if x==7:
        return 3
    if x==8:
        return 4
    if x==9:
        return 4
    if x==10:
        return 9    
    if x==11:
        return 9   
    if x==12:
        return 10
    if x==13:
        return 10
    if x==14:
        return 10  
    if x==15:
        return 2  
    if x==16:
        return 2
    if x==17:
        return 2
    if x>=18:
        return 4
```

```python
df['waveR']=df['wave'].apply(score_to_numeric)
df.head()
```

```python
# Feb-Apr recode = 1; Sep-Nov recode = 2
def month_to_season(x):
    if x<=4:
        return 1
    if x>=9:
        return 2
```

```python
df['season']=df['waveR'].apply(month_to_season)
df.head()
```

```python
# QA recode to ensure recodes "wave to month" are correct
QA_A= pd.crosstab(index=df["wave"], columns=df["waveR"])
QA_A
```

```python
# QA recode to ensure recodes "month to season" are correct
QA_B= pd.crosstab(index=df["waveR"], columns=df["season"])
QA_B
```
